```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/kotagiri-preeti/Summer-Internship-AI-and-ML/master/Match%20(1).csv")
df
```

```python
#1

from collections import Counter
manofmatch=Counter(df['ManOfMach'])
count=pd.DataFrame(list(manofmatch.items()),columns=["players","manofmatch"])
count.sort_values("manofmatch",ascending=False)
count.sort_values("players",ascending=True)
```

```python
#2


country=Counter(df['Country_Name'])
a=list(country.keys())
b=list(country.values())

import matplotlib.pyplot as plt
%matplotlib inline

plt.pie(b,labels=a,autopct='%0.2f%%',explode=[0.2,0,0],radius=1.5,startangle=45,colors=['b','r','g'])
plt.show()
```
